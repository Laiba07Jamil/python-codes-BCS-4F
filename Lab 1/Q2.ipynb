```python
#Question 2: Even Number Counter
#Question:
#Write a Python program that takes a number n from the user and prints all even numbers from 1
#to n. Also display the total count of even numbers.

n = int(input("Enter a number: "))
i = 1
count = 0

print("Even numbers: " , end = " ")
while i <= n:
    if(i % 2 == 0):
        print(i , end = " ")
        count = count + 1
    i = i + 1
print(" ")
print("Total Even numbers: ", count)
```
