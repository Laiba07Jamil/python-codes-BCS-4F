```python
#Write a Python program that takes a student’s name and marks as input and prints the grade
#according to the following criteria:
#• Marks ≥ 85 → Grade A
#• Marks ≥ 70 → Grade B
#• Marks ≥ 50 → Grade C
#• Marks < 50 → Fail

name = input("Enter student name: ")
marks = int(input("Enter marks: "))

print("Student Name: " , name)
print("Marks: ",marks)

if(marks >= 85):
    print("Grade A")

elif (marks >= 70 and marks <= 84) :
    print("Grade B")

elif(marks >= 50 and marks  <= 69):
    print("Grade C")
else :
    print("Fail")
```
