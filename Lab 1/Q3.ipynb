```python
# Question 3: Student Record Using Dictionary
#Question:
#Write a Python program that stores the names and marks of three students in a dictionary and
#then displays all student records.

Records = {}

for i in range(3):
    name = input("Enter the name: ")
    marks = int(input("Enter marks: "))
    Records.update({name : marks})

print("Student Records: ")
for name , marks in Records.items():
    print(name , " : " , marks)

```
