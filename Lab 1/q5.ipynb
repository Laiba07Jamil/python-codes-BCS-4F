```python
#Question 5: Average Marks Using Function
#Question:
#Write a Python function named calculate_average() that takes a list of marks and returns the
#average marks.

def calculate_average(list1):
      total = 0
      for m in list1:
        total += m
      return total / len(list1)


list1 =[]
n = int(input("Enter the number of marks: "))
for i in range(0,n):
      ele = int(input("Enter marks: "))
      list1.append(ele)

result = calculate_average(list1)
print("Marks: ",list1)
print("Average Marks: ",result)

```
